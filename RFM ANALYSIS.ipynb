```python
##Recency of each customer
```

```python
df_recency = df.groupby(by = 'Customer Name', as_index = False) ['Order Date'].max()
df_recency.columns = ['CustomerName', 'LastPurchaseDate']
recent_date = df_recency['LastPurchaseDate'].max()
```

```python
df_recency.head()
```

```python
df_recency['Recency']= df_recency['LastPurchaseDate'].apply(lambda x:(recent_date - x).days)
df_recency.head()
```

```python
frequency_df= df.drop_duplicates().groupby(by = ['Customer Name'],as_index = False) ['Order Date'].count()
frequency_df.columns = ['CustomerName', 'Frequency']
frequency_df.head()
```

```python
df['Total'] = df['Sales'] * df['Quantity']
monetary_df = df.groupby(by = ['Customer Name'], as_index = False) ['Total'].sum()
monetary_df.columns = ['CustomerName', 'Monetary']
monetary_df.head()
```

```python
##Creating RFM table
```

```python
rf_df = df_recency.merge(frequency_df, on='CustomerName')
rfm_df = rf_df.merge(monetary_df, on='CustomerName').drop(columns='LastPurchaseDate')
rfm_df.head()
```

```python
##Creating rank for recency, frequency and monetary
```

```python
rfm_df['R_rank']= rfm_df['Recency'].rank(ascending= False)
rfm_df['F_rank']= rfm_df['Frequency'].rank(ascending= True)
rfm_df['M_rank']= rfm_df['Monetary'].rank(ascending= True)

rfm_df['R_rank_norm']= (rfm_df['R_rank']/rfm_df['R_rank'].max())*100
rfm_df['F_rank_norm']= (rfm_df['F_rank']/rfm_df['F_rank'].max())*100
rfm_df['M_rank_norm']= (rfm_df['M_rank']/rfm_df['M_rank'].max())*100

rfm_df.drop(columns= ['R_rank', 'F_rank', 'M_rank'], inplace= True)
rfm_df.head()
```

```python
##using above normalized columns now we are going to calculate RFM SCORE.
```

```python
rfm_df['RFM_Score']= 0.15* rfm_df['R_rank_norm']+ 0.28* rfm_df['F_rank_norm']+ 0.57* rfm_df['M_rank_norm']
rfm_df['RFM_Score']*=0.05
rfm_df= rfm_df.round(2)
rfm_df[['CustomerName', 'RFM_Score']].head()
```

```python
rfm_df['RFM_Score'].describe()
```

```python
##Categorize columns based on the scores.
#->4.5 TOP CUSTOMERS
#3-4 MEDIUM VALUE CUSTOMERS
#1.5-3 LOW VALUE CUSTOMERS
#<=1.5 LOST CUSTOMERS
```

```python
rfm_df["Customer_segment"] = np.where(rfm_df['RFM_Score']>4.5, "Top Customer",

                             (np.where(rfm_df['RFM_Score']> 4, "High Value Customer",

                                       (np.where(rfm_df['RFM_Score'] > 3, "Medium Value Customer",

                                                 np.where(rfm_df['RFM_Score']> 1.5,

                                                          "Low Value Customer", "Lost Customers"))))))

rfm_df[['CustomerName', 'RFM_Score', 'Customer_segment']].head(20)
```

```python
##Pie-chart showing customer segmentation
```

```python
plt.pie(rfm_df.Customer_segment.value_counts(),
        labels= rfm_df.Customer_segment.value_counts().index,
        autopct= "%.0f%%")
plt.show()
```

```python
##From this above pie-chart we can interepret customer segmentation based on their value and RFM Score
##Lower the score higher is the value of the customer and higher the score lower is the value of that customer.
##Based on this information we will follow marketing strategies, target marketing, or incentives or discounts etc.
```
